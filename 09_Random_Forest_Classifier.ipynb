```python
input_data = "data/data_random_forests.txt"
```

```python
with open(input_data) as f:
    print(f.readline())
```

```python
import numpy as np
data = np.loadtxt(input_data,delimiter = ',')
X,y = data[:,:-1], data[:,-1]
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators= 100, max_depth= 4, random_state= 0)
```

```python
model.fit(X,y)
```

```python
y_pred = model.predict(X)
```
